# Creating a `pandas.DataFrame`

*Many ways of creating a dataframe*. Let's see one. 
Remeber dictionaries? They are sets of key-value pairs:

```
{ 'key1' : value1, 
  'key2' : value2
}
```

Think of a 'pd.DataFrame' as a dictionary where: 
- keys = column names
- values = column values

We can create a dataframe like this:

````python
# initialize dictionary with columns'data
d={'col_name_1':np.arange(3),
  'col_name_2' : [3.1, 3.2, 3.3]
  }
d
````

````python
# create data frame
df = pd.DataFrame(d)
df
````

# In-place operations
Let's rename the data frame's column. 
We cab use a dataframe *method* called `rename`.
`rename` takes in as an input a dictionary: 

```
{'col_1_old_name' : 'col_1_new_name',
 'col_2_old_name' : 'col_2_new_name'}

```

````python
# define new column names
col_names = {'col_name_1' : 'col1',
             'col_name_2' : 'col2'}

# rename using rename
df.rename(columns = col_names)
````

````python
# take a look at our dataframe
df
````

Nothing changed: 
`df.rename()` doesn't change the column names *in place*, meaning it doesn't modify the object itself. Instead, it created a new object as an output.

Assign output back to dataframe ot actually change it: 

````python
df=df.rename(columns = col_names)
df
````

## Selecting a single column by column name
Simplest case for selecting data. The syntax is : 
    
```
df[`column_name`] 
```

NOTE: this is the same syntax for a dictionary. Remeber we can think of a pd.DataFrame as a dictionary where the keys are the column names. 

Example: we are interested in sea level rise. 

````python
# select a single column using []
annual_rise = df['annual_sea_level_rise']

# check the type of the output
print(type(annual_rise))

annual_rise
````

`df['col_name']` is an example of selecting **selecting by label** means we select data from the data frame using the names of the columns, not their position.

# Selecting nultiple columns using a list of column names
The syntax for selecting nultiple columns using a list of column names is 
```
df[['col1','col2','col3']]
```

The list of column names `['col1', 'col2', 'col3']` goes inside the selection brakets `[]`. That's why there are double brackets. 

Notice each column name is a string.

Example: select change in glacial volume in Europe and Asia. 

````python
# select columns with names "europe" and "asia"
europe_asia = df[['europe','asia']]

#print type of resulting selection
print(type(europe_asia))

# check the shape of the selection
print(europe_asia.shape)

# see our selection
europe_asia
````

## Selecting nultiple columns using a slice

We will use `loc` selection. The gneeral syntax for `loc` is 
```
df.loc[row-selection, column-selection]
```

where
- `row-selection` = rows we want to subset from `df`
- `column-selection` = columns we want to subset from `df`.

Examples: select change in galcial volumne per year in all regions.

````python
df.head()
````

````python
# select all columns between europe and antarctica 
df.loc[:, 'europe':'antarctica'].head()
````

Notes about this: 
- `row-selection` = `:` colon = means select all the rows
- `column-selection` = slice `'europe:antarctica'`  
- we get both ends of the slice in the selected dataframe. Different from slicing in base Python and Numpy, where the endpoint is *not* included in the slice.

# Select rows

## Selecting rows using a condition 

General syntax: 

    ```
    df[condition_on_rows]
    ```
Example: we are interested in data after 1996


````python
# select all rows with year > 1996
df[df['year']>1996]
````

Let's break it down: 
- in the `df[conditional_on_rows]` syntax, we have that `condition_on_rows` = `df['year'] > 1996`
- `df['year]' > 1996` checks which rows have values greater than 1996 i the year column. Let's see: 


````python
df['year'] > 1996
````

- `df['year'] >1996` is a `pd.Series` with boolean values (True or False)
- `pd.Series` with boolean values (True of False) are often called **masks**
- when we pass a `pd.Series` with boolean vlaues(mask) to the selection brackets[]. we keep only those rows with a `True` value in the pd.Series/mask

Example : select data from years 1970 to 1979 (including 1979)

````python
range(1970, 1980)
````

````python
df[df.year.isin(range(1970,1980))]
````

## Select rows using multiple conditions
We can combine multiple conditions by surroudning each one in parameter is `()`, and using the **or operator** `|` and the ** and operator** `&` 

The syntax is : 
    
```
# select rows of df that satisfy conditional OR condition 2 
df[ (condition1) | (condition2)]
```
and 

```
# select rows of df that satisfy condtion1 AND condition2
df[ (condition1) & (condition2)]
```
Example:

````python
# select rows with
# annual sea level rsie < 0mm OR annual sea level rise > 0.8mm

df[(df['annual_sea_level_rise'] < 0) | (df['annual_sea_level_rise']>0.8)]
````

## Selecting rows by position
Somties we want to select rows on their *actual* position in the data frame. 
Use `iloc` selection, syntax is : 
```
df.iloc[row-indices]
```

`iloc` = stands for integerlocation based indexing.

Example: 

````python
# select the fifth row = index 4
df.iloc[4]
````

````python
# select rows 23 through 30, including 30
df.iloc[23:31]
````

# Selecting rows and columns simultaneously 
We can do this using `loc` (for labels or conditions) or `iloc` (integer positions). 

# Rows and columns by labels or conditions
We use `loc`:
```
df.loc[ row-selection , column-selection]
```
`row-selection` and `column-selection` can be a condition or subset of labels from the index or the column names.

Example: 

````python
# select change in glacial volume in Europe per year after 2000
df.loc[df['year'] > 2000, ['year','europe']]
````

````python
## Selecting rows and columns by position
we use `iloc` with the syntax

```
df.iloc[ row-indices, columns-indices]
```

Example: 
````

````python
# select rows with indices 3-7 (including 7) and the 4th and 5th column.
df.iloc[3:8,3:5]
````
