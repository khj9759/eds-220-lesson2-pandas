# `plot()` method
`pandas.DataFrame` has a built-up method `plot()` for plotting.
Default behavior of plot without specifying parameters: 

- creates one line plot for each of the columns with numeric data
- sets values of the index as the x-axis values

Example: 

````python
# a mess
df.plot()
````

## Line plots
We can make a line plot of one column against another by using the following syntax:
```
df.plot(x='x_values_column', y=''y_values_column')
```
Example: 

````python
# change in glacial volume per year in Europe
df.plot(x='year', y='europe')
````

Basic customization specifying other arguments in `plot()`.

Example: 

````python
# change in glacial volume per year in Europe
df.plot(x='year',
        y='europe',
        title='Change in galcial volume per year in Europe',
       xlabel = "'year',
       ylabel = 'change in glacial volume (km2)',
       color='green')
````

#Multiple line plots
we want to graph the change in glacial volume in the Arctic and Alaska, update these arguments: 
- 'y': a list of column names that will be plotted agains x
- 'color': specify the color of each column with a dictionary `{'col1':'color1', 'col2':'color2'}`

Example:

````python
colors = {'arctic':'#F48FB1',
         'alaska':'yellow'}
df.plot(x='year',
       y=['arctic', 'alaska'],
       color = colors)
````

Create seperate plots setting the `subset` argument `True`

````python
df.plot(x='year',
       y=['arctic', 'alaska'],
       color = colors,
       subplots=True)
````

# `kind()` argument in `plot()`
`plot()` default = creating a line plot.
parameter that controls what kind of plot we get is `kind` Let's see in detail

# Scatter plots
Example: visualy compare the flipper length against body mass. 

````python
penguins.plot(kind='scatter',
             x='flipper_length_mm',
             y='body_mass_g',
             alpha=0.4 #controls transparency of points
             )
````

## Bar plot
Set `kind`=`bar` in the `plot()` methods.
Example: we want to get data about the 10 penguins with the lowest body mass and visualize it. 

1. Find penguins with smalles body mass using `nsmallest()`method for series.



````python
smallest = penguins.body_mass_g.nsmallest(5)
smallest
````

````python
#sort_values orders values in descending order
smallest = penguins.body_mass_g.nsmallest(5).sort_values(ascending=False)
smallest
````

````python
#plot smallest body masses as bars
smallest.plot(kind='bar')
````

What if we wanted to look at other data for these smallest penguins?
We can use the `index` of the `smallest` series to select those rows in the original `penguins` data frame(use `loc`)

````python
penguins.loc[smallest.index] # index already sorted.
````

# Histogram
To create a histrogram, set `kind`=`hist` in `plot()`

````python
# using plot without subsetting data - a mess again
penguins.plot(kind='hist')
````

````python
# distributino of filpper length measurement
# first select data, then plot
penguins.flipper_length_mm.plot(kind='hist',
                               title='Penguins flipper length distribution',
                               xlabel='Flipperlength (mm)',
                               grid=True)
````
